```python
import matplotlib.pyplot as plt

# numerical column
numerical_features = [
    'danceability', 'energy', 'valence', 'tempo', 'loudness',
    'acousticness', 'instrumentalness', 'speechiness', 'duration_ms'
]

# visualization - matplotlib
df_high[numerical_features].hist(bins=20, figsize=(14,10), color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Numerical Features', fontsize=16)
plt.show()
```

```python
# visualization - seaborn
import seaborn as sns

sns.set(style="whitegrid", palette="pastel", font_scale=1.1)

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(numerical_features):
    sns.histplot(df_high[col], bins=30, kde=True, ax=axes[i], color=sns.color_palette("Set2")[i % 8])
    axes[i].set_title(f"Distribution of {col}", fontsize=12)
    axes[i].set_xlabel("")
    
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.suptitle("Seaborn Histograms of Numerical Features", fontsize=16, y=1.02)
plt.show()
```

```python
# Caculate Correlation

corr = df_high[numerical_features].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features', fontsize=16)
plt.show()
```

```python
# -----------------------------------------------여기서 부터 다시 개발 (10/28 화)-----------------------------------------------------
# Select Feature

feature_cols = [
    'danceability', 'energy', 'tempo'
]

X = df_high[feature_cols].values
X
```

```python
# User Vector Example

user_history = ['2plbrEY59IikOBgBGLjaoe', '6dOtVTDdiauQNBQEDOtlAB', '7ne4VBA60CxGM75vw0EYad']
user_songs = df_high[df_high['track_id'].isin(user_history)][feature_cols]
user_vector = user_songs.mean().values
```

```python
from sklearn.neighbors import NearestNeighbors

# KNN Model
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(X)

# Recommend 10 songs
distances, indices = knn.kneighbors(user_vector.reshape(1, -1), n_neighbors=10)

# Result
recommended_songs = df_high.iloc[indices[0]][['track_name', 'track_artist', 'playlist_genre']]
```

```python
from IPython.display import display

display(
    recommended_songs[['track_name', 'track_artist', 'playlist_genre']]
        .rename(columns={
            'track_name': '🎵 Track Name',
            'track_artist': '🎤 Artist',
            'playlist_genre': '🎧 Genre'
        })
        .style.set_properties(**{
            'background-color': '#fafafa',
            'border-color': '#cccccc',
            'text-align': 'center'
        })
)
```
